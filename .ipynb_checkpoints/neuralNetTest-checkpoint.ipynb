```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
class ANN:
    def __init__(self, layers):
        if len(layers) < 2:
            raise Exception('Please have a valid network shape.')
        
        self.layers = layers
        
        self.weights = []
        self.biases = []
        self.parse_layers(layers)
    
    def parse_layers(self, layers):
        for i in range(len(layers)-1):
            self.weights.append(np.random.rand(layers[i+1], layers[i]))
            self.biases.append(np.random.rand(layers[i+1]).squeeze())
            
    def transfer_func(self, x):
        return x
        if x > 1:
            return 1
        if x < -1:
            return -1

    
    def get_output(self, input):
        if len(input) != self.layers[0]:
            raise Exception('Please use a valid input shape.')
        
        for i in range(len(self.weights)):
            input = np.array(list(map(self.transfer_func, np.matmul(self.weights[i], input) + self.biases[i])))
        
        return input
```

```python
net = ANN((1, 5, 6, 1))

x = np.linspace(-100, 100, 1000)

y = np.array([])

for i in x:
    y = np.append(y, net.get_output(np.array([i])))

plt.plot(x, y)
```
