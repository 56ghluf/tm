### Use the iteration methode to approximate functions 

```python
# Imports
import numpy as np
from time import time
import matplotlib.pyplot as plt
```

```python
# Define a funciton to approximate
def f(x):
    return np.sqrt(3*x) + 1
```

```python
# Define a function to cacluate an interval
# over which there is an intersection

def find_interval(f, a, timeout, step):
    b = a + step
    c_1 = f(a) - a
    t_1 = time()
    while not(c_1 * (f(b)-b) < 0) and time() - t_1 < timeout:
        b += step
    return b
```

```python
# Caclulate an interval for the function
a = 0
b = find_interval(f, a, 10, 0.1)
```

```python
b
```

```python
# Define a funciton to approximate the intersection
def approx(f, a, b, n):
    xs = [(a+b)/2]
    for i in range(n):
        xs.append(f(xs[len(xs)-1]))
    return xs
```

```python
# Approximate the zero
xs = approx(f, a, b, 10)
```

```python
xs
```

```python
# Function to generate point pairs to plot segments
def gen_point_pairs(f, ps):
    x_plots, y_plots = [], []
    
    for i in range(len(ps)-1):
        x_plots.append([ps[i], ps[i]])
        y_plots.append([ps[i], f(ps[i])])
        
        x_plots.append([ps[i], ps[i+1]])
        y_plots.append([f(ps[i]), ps[i+1]])

        
    return x_plots, y_plots
```

```python
# Generate the point pairs
x_pairs, y_pairs = gen_point_pairs(f, xs)
```

```python
# Graph the function
plt.figure(figsize=(10, 10))

# Define an array of the interval to graph
I = np.linspace(a, b, 100)

# Plot the function
plt.plot(I, f(I))

# Plot y=x
plt.plot(I, I)

# Plot all the points that were used to approximate
for i in range(len(x_pairs)):
    plt.plot(x_pairs[i], y_pairs[i], 'r')
```
